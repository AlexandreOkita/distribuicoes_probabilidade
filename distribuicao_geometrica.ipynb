<h1> Distribuição geometrica </h1>

<h3>Este notebook possui o proposito de desenhar graficos para ilustrar uma distribuição geométrica </h3> 

<div class="alert alert-block alert-info">
<b>Formula utilizada:</b> (1-p)^(k-1) * p
</div>

```python
def encontrar_prob(k, p):
    return (1-p)**(k-1) * p
```

```python
import pandas
probs = []

for k in range(1, 10):
    probs.append(encontrar_prob(k, 0.2))

Probs = pandas.DataFrame({'Probabilidade de vencer em n vezes': probs})
Probs.plot.line()
```

```python
probs_acum = []
probs_acum.append(probs[0])
for x in range(1, len(probs)):
    probs_acum.append(probs[x] + probs_acum[x-1])
Probs_acum = pandas.DataFrame({'Probabilidade de vencer em até n vezes': probs_acum})
Probs_acum.plot.line()
```
